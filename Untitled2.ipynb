# Text_preprocessing

```python
def text_lowercase(text): 
    return text.lower()
def cleanHtml(sentence):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, ' ', str(sentence))
    return cleantext


def cleanPunc(sentence): #function to clean the word of any punctuation or special characters
    cleaned = re.sub(r'[?|!|\'|"|#]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    cleaned = cleaned.strip()
    cleaned = cleaned.replace("\n"," ")
    return cleaned


def keepAlpha(sentence):
    alpha_sent = ""
    for word in sentence.split():
        alpha_word = re.sub('[^a-z A-Z]+', ' ', word)
        alpha_sent += alpha_word
        alpha_sent += " "
    alpha_sent = alpha_sent.strip()
    return alpha_sent
```

```python
train_data1['text_col']=train_data1['text_col'].apply(text_lowercase)
train_data1['text_col']=train_data1['text_col'].apply(cleanHtml)
train_data1['text_col']=train_data1['text_col'].apply(cleanPunc)
train_data1['text_col']=train_data1['text_col'].apply(keepAlpha)

test_data1['text_col']=test_data1['text_col'].apply(text_lowercase)
test_data1['text_col']=test_data1['text_col'].apply(cleanHtml)
test_data1['text_col']=test_data1['text_col'].apply(cleanPunc)
test_data1['text_col']=test_data1['text_col'].apply(keepAlpha)
```

```python
stop_words = set(stopwords.words('english'))
stop_words.update(['zero','one','two','three','four','five','six','seven','eight','nine','ten','may','also','across','among','beside','however','yet','within'])
re_stop_words = re.compile(r"\b(" + "|".join(stop_words) + ")\\W", re.I)
def removeStopWords(sentence):
    global re_stop_words
    return re_stop_words.sub(" ", sentence)

train_data1['text_col'] = train_data1['text_col'].apply(removeStopWords)
test_data1['text_col'] = test_data1['text_col'].apply(removeStopWords)
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

train_data1['text_col']=train_data1['text_col'].apply(lambda x: ' '.join(
    wordnet_lemmatizer.lemmatize(term) for term in x.split()))

test_data1['text_col']=test_data1['text_col'].apply(lambda x: ' '.join(
    wordnet_lemmatizer.lemmatize(term) for term in x.split()))
```

```python
# Remove word stems using a Porter stemmer
ps = nltk.PorterStemmer()

train_data1['text_col'] = train_data1['text_col'].apply(lambda x: ' '.join(
    ps.stem(term) for term in x.split()))
test_data1['text_col'] = test_data1['text_col'].apply(lambda x: ' '.join(
    ps.stem(term) for term in x.split()))
```

```python
train_data1['Computer Science']=train_data1['Computer Science'].astype(int)
train_data1['Physics']=train_data1['Physics'].astype(int)
train_data1['Mathematics']=train_data1['Mathematics'].astype(int)
train_data1['Statistics']=train_data1['Statistics'].astype(int)
train_data1['Quantitative Biology']=train_data1['Quantitative Biology'].astype(int)
train_data1['Quantitative Finance']=train_data1['Quantitative Finance'].astype(int)
```
